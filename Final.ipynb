# 4. Background and/or Related Work

The emergence of the COVID-19 pandemic aggravated an already existing public mental health problem. As a consequence of COVID-19, the pandemic created an environment exacerbating the determinants of poor mental health. According to the Anxiety & Depression Association of America (ADAA), [40 million American adults aged 18 and older](https://adaa.org/understanding-anxiety/facts-statistics#:~:text=GAD%2520affects%25206.8%2520million%2520adults,to%2520be%2520affected%2520as%2520men.) suffer anxiety disorders; the most common mental illness in the United States (US). Fear and grief associated with disease spread, social restrictions, and other pandemic containment measures to reduce mortality and morbidity worsen psychological wellbeing because of feelings of isolation and even loss of livelihoods and loved ones. Unsurprisingly, the percentage of individuals with anxiety or a depressive disorder symptom increased from [36.4% to 41.5%](https://www.cdc.gov/mmwr/volumes/70/wr/mm7013e2.htm#suggestedcitation) during the August 2020-to-February 2021 period. With the reality of worsening situations related to mental health, understanding the trends of anxiety and depression is crucial in guiding health and public policy formulation. As the burden of psychological disorders continues to plague the healthcare system, there is a need for better risk assessment as well as intervention and prevention.

# 6. Using Data Analysis to answer Research Questions

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Loading and Inspection

```python
data = pd.read_csv('depression_indicators.csv')
```

```python
data['Group'].unique()
```

```python
data.info()
```

### Data Preparation

```python
data['Time Period Start Date'] = data['Time Period Start Date'].astype('datetime64')
data1 = data.set_index(['Time Period Start Date'] )

data['Time Period End Date'] = data['Time Period End Date'].astype('datetime64')
data2 = data.set_index(['Time Period End Date'] )
```

```python
depression1 = data1.loc[data1['Indicator'] == 'Symptoms of Depressive Disorder']
anxiety1 = data1.loc[data1['Indicator'] == 'Symptoms of Anxiety Disorder']
```

```python
depression2 = data2.loc[data2['Indicator'] == 'Symptoms of Depressive Disorder']
anxiety2 = data2.loc[data2['Indicator'] == 'Symptoms of Anxiety Disorder']
```

### Inspecting research Question 1.

```python
plt.figure(figsize = (10,5))
sns.lineplot(data = depression1[ 'Value'], color = 'green')
sns.lineplot(data = depression2[ 'Value'], color = 'red')
sns.lineplot(data = anxiety1['Value'], color = 'brown')
sns.lineplot(data = anxiety2['Value'], color = 'blue')
plt.legend(['Depression at Start','','Depression at End','', 'Anxiety at Start','','Anxiety at End', ''])
plt.xticks(rotation = 30)
plt.title('Anxiety and Depression Values in Time', fontsize = 20)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Levels of Anxiety or Depression', fontsize = 15)
```

### We saw that trend of depression level and trend of anxiety level is similar.


### Inspecting research Question 2.

```python
q2_data = data[['Indicator',
                      'Group',
                      'Subgroup',
                      'Time Period Start Date','Time Period End Date',
                      'Value']]
```

```python
q2_depr = q2_data.loc[(q2_data['Group'] == 'By Age') & (q2_data['Indicator'] == 'Symptoms of Depressive Disorder')]
q2_anx = q2_data.loc[(q2_data['Group'] == 'By Age') & (q2_data['Indicator'] == 'Symptoms of Anxiety Disorder')]
```

```python
depression_by_age = q2_data.groupby(['Subgroup']).mean()
anxiety_by_age = q2_anx.groupby(['Subgroup']).mean()

by_age = depression_by_age.merge(anxiety_by_age, left_index = True, right_index = True)
by_age.rename(columns = {'Value_x' : 'Depression Level', 'Value_y' : 'Anxiety Level'}, inplace = True)

by_age
```

```python
by_age.plot(kind = 'bar', rot = 0, figsize = (10,5))
plt.xlabel('Age Group')
plt.ylabel('Average Depression or Anxiety Level')
plt.title('Average Depression and Anxiety by Age Group')
```

### Inspecting research Question 3.

We got the data for trend of covid 19 cases from the CDC website: https://covid.cdc.gov/covid-data-tracker/#trends_dailycases

```python
covid_data = pd.read_csv("covid19trend.csv")
covid_data.head()
```

```python
covid_data.columns
```

```python
covid_data['Date'] = covid_data['Date'].astype('datetime64')
covid_data.set_index(['Date'], inplace = True)

covid_data.drop(['Historic Cases'],axis = 1,  inplace = True)
```

```python
#Scaling datasets to show no. of cases and depression/anxiety levels in comparable scale
mapped = depression1[['Value']]/max(depression1['Value'])
mapped_1 = anxiety1[['Value']]/max(anxiety1['Value'])
covid_mapped = covid_data[['7-Day Moving Avg']]/max(covid_data['7-Day Moving Avg'])
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(data = covid_mapped['7-Day Moving Avg'] , color = 'green')
sns.lineplot(data = mapped['Value'], color = 'red')
sns.lineplot(data = mapped_1['Value'], color = 'orange')
plt.grid()
plt.legend(['Cases', 'Depression Value','', 'Anxiety Value'])
plt.ylabel('Average Covid Cases or Depression/Anxiety Value')
plt.title('Trend of Covid Cases, Depression Levels and Anxiety Levels')
```

### Inspecting research Question 4.

```python
depression1['Group'].unique()
```

```python
dep_by_race = depression1.loc[depression1['Group'] == 'By Race/Hispanic ethnicity']
anx_by_race = anxiety1.loc[anxiety1['Group'] == 'By Race/Hispanic ethnicity']
```

```python
races = dep_by_race['Subgroup'].unique()
colors = ['red', 'brown', 'green', 'blue', 'orange']
```

```python
plt.figure(figsize = (14,8))
for race, color in zip(races, colors):
    dep = dep_by_race.loc[dep_by_race['Subgroup'] == race]
    sns.lineplot(data = dep['Value'], color = color)
plt.legend(races)
plt.title('Depression trend for various races', fontsize = 15)
plt.xlabel('Time Period', fontsize = 12)
plt.ylabel('Depression Level', fontsize = 12)
```

```python
dep_bar = dep_by_race[['Subgroup','Value']].groupby(by = 'Subgroup').mean()

dep_bar.plot(kind = 'bar',
             rot = 15,
             figsize = (12,7),
             ylabel = 'Average recorded depression level')
plt.title('Average Depression Levels of Different Race', fontsize = 15)
plt.xlabel('Race/Ethnicity')
```

```python
plt.figure(figsize = (14,8))
for race, color in zip(races, colors):
    anx = anx_by_race.loc[anx_by_race['Subgroup'] == race]
    sns.lineplot(data = anx['Value'], color = color)
plt.legend(races)
plt.title('Anxiety trend for various races', fontsize = 15)
plt.xlabel('Time Period', fontsize = 12)
plt.ylabel('Anxiety Level', fontsize = 12)
```

```python
anx_bar = anx_by_race[['Subgroup','Value']].groupby(by = 'Subgroup').mean()

anx_bar.plot(kind = 'bar',
             rot = 15,
             figsize = (12,7),
             ylabel = 'Average recorded anxiety level')
plt.title('Average Anxiety Levels of Different Race', fontsize = 15)
plt.xlabel('Race/Ethnicity')
```

### Inspecting research Question 5.
